```python
def urlparser(url : list):
    """
    This function accepts a URL and returns the parsed components of the URL
    parm url : Input parameter is url 
    returns  : A dict which contains all the parsed components of the URL
    """
    global parsed_components
    parsed_components = {}  # dictionary
    
    # URL conatins the following parts 
    # scheme, netloc, path, params, query, fragment. The below line shows the representation of each part in the URL.
    # "scheme://netloc/path;parameters?query#fragment" 
    
    
    # Schema part is before the '//'
    def get_schema(double_slash_pos):
        return url[:double_slash_pos - 1]
    
    def get_netloc(double_slash_pos,single_slash_pos):        
        network_location = url[double_slash_pos + 2 : double_slash_pos + 2 + single_slash_pos ]
        #network_location is the socket which contains both IP and Port number.
        if len(network_location.split(':')) == 2: 
            return network_location
        
        return ''
    
    def get_path(schema,netloc):
        
       
        #Checking if the schema is present.
        if len(schema) > 0:
            url_stripped_schema = url.replace(schema+'://','')
            #print("Schema is ",schema)
            #print(url_stripped_schema)
        
        elif len(schema) == 0 and len(netloc) > 0:
            #Here url is starting with network location
            url_stripped_schema = url.replace('//','')
           
        if len(netloc) > 0:
            url_stripped_netloc = url_stripped_schema.replace(netloc,'')
            #print("Net loc is",netloc)
            #print(url_stripped_netloc)
        else :
            url_stripped_netloc = url
        
        parts = url_stripped_netloc.split(';')
        return parts[0]

    def get_fragment():
        try:
            return url.split('#')[1]
        except IndexError:
            return ''
    
    def get_query():
        try:
            return url.split('?')[1].split('#')[0]
        except IndexError:
            return ''
    
    double_slash_pos = url.find("//")
    #print("// is at ",double_slash_pos )
    single_slash_pos = url[double_slash_pos + 2 :].find('/')
    #print("/ is at ",single_slash_pos)
    semicolon_pos = url.find(';')

    if double_slash_pos > 0 :
        
        parsed_components['schema'] = get_schema(double_slash_pos)
        parsed_components['netloc'] = get_netloc(double_slash_pos,single_slash_pos)
        
    elif double_slash_pos == 0 :
        #when '//' is at the first position of URL
        parsed_components['schema'] = ''
        parsed_components['netloc'] = get_netloc(double_slash_pos,single_slash_pos)

    elif double_slash_pos == -1: 
        #when // is not found in the URL
        parsed_components['schema'] = ''
        parsed_components['netloc'] = ''

    parsed_components['path'] = get_path(parsed_components['schema'],parsed_components['netloc'])
    parsed_components['query'] = get_query()
    parsed_components['fragment'] = get_fragment()

    
    #print(parsed_components)
    return parsed_components
```
