```python
def prob_generative_model(wordsLen_population:list ,wordsLen_Prob:list,chars_population:list,chars_Prob:list)->str:
    """
    This function will generate a word from the given distribution of word length and characters frequencies 
    
    :param wordsLen_population : Input parameter which contians a length of words from the observerd distrivution(text file) 
    :param wordsLen_Prob       : Input parameter which contians probabilities of the length of each word
    :param chars_population    : Input parameter which contians set of characters from the observerd distrivution(text file) 
    :param chars_Prob          : Input parameter which contians probabilities of each character
    
    return word
    """
    
    # Selcting the word length of the word to be generated 
    # using numpy.random() and passing the observerd wordlength distribution and its corresponding probabilites
    word_length = np.random.choice(wordsLen_population, p=wordsLen_Prob)
    
    #Forming a word of the above length from the observerd Character distribution and its corresponding probabilites
    word = ""
    word = "".join([word + np.random.choice(chars_population, p=chars_Prob) for _ in range(word_length)])
    
    return word
    
    
```

```python
# Generating 5000 words 
generated_words = []
for i in range(5000):
    word = prob_generative_model(wordsLen_population,wordsLen_Prob,chars_population,chars_Prob)
    generated_words.append(word)
    
```

 # Modelling Web as Graph

```python
def get_adjacency(filename)->np.ndarray:
    """
    This function will create the adjacency matrix for the given undirected graph which is stored in the .dat file
    in the form of edges.
    
    :param filename : Input paramter which represents the filename for the .dat file
    
    returns : numpy array which represents the adjacency matrix
    """
    
    

    try :
        # reading the file  and creating the file object     
        f = open(filename)
        lines = []
        for line in f.readlines():
            lines.append(line)
        #print(lines)
        '''
        Here line 1 says 'sym unweighted' means given graph is symmetric and unweighted
        10 means number of undirceted edges for the given graph
        8 represents the number of vertices   
        (for clear understanding in the file structure use print(lines))
        '''    
        
        nodes = int(lines[1].split()[3])
        #print(nodes)
        adjacency_matrix = np.zeros((nodes,nodes))
        for i in range(2,len(lines)):
            x,y = lines[i].split()
            #print(x,y)
            adjacency_matrix[int(x)-1][int(y)-1] = 1
            adjacency_matrix[int(y)-1][int(x)-1] = 1
                    
    finally:
        #closing the file
        f.close()
        
    return  adjacency_matrix
        
```

### GINI COEFFICIENT FOR THE ADJACENT MATRIX

```python
def gini_coeff(adj_matrix:np.ndarray)->float:
    """
    This function will calculate the gini coefficient for the given undirected graph
    
    :param adj_matrix : input parameter which is an numpy array which represents the Adjacenct matrix of the undirected graph
    
    :returns : float value which is the gini coefficient
    """
    # Calculating the degree for the vertices from the given adjacent matrix
    # Degree(node) = Indegree(node) + Outdegree(node)
    
    degrees = np.sum(adjacency_matrix,axis=1)
    
    #Sorting the degrees
    degrees.sort()
    n = len(degrees) 
    numerator = 0
    for i in range(n):
        for j in range(n):
            temp = np.abs(degrees[i]-degrees[j])
            numerator = numerator + temp
            
    denominator = 2*n*np.sum(degrees)
    gini = numerator/denominator
    
    return gini

```

Find the traingles in the adjacent graph

```python

def count_triangles(adj_matrix):
    """
    This function will find the number of traingles in the graph
    
    :param adj_matrix : input parameter which is an numpy array which represents the Adjacenct matrix of the undirected graph

    returns : number of traingles
    """
    

    
    vertices_list = []
    traingles = []
    
    num_vertices = len(adj_matrix)
    
    # Iterate through each vertex in the graph
    for i in range(num_vertices):

        # Getting neighbors for the current vertex
        neighbors = [j for j in range(num_vertices) if adj_matrix[i][j] == 1]
        #print("i=",i)
        #print(neighbors)
  
        # Checking for triangles 
        for j in neighbors:
            for k in neighbors:
                # If vertex j and vertex k are connected and j and k are not the same vertex,
                # then we have found a triangle
                if adj_matrix[j][k] == 1 and j != k:
                    #print("j=",j)
                    #print("k=",k)
                    #print(count)
                    vertices_list.append([i,j,k])
                    
                    
    vertices_list = [set(i) for i in vertices_list ]                    
    #print(vertices_list)     
    
    for i in range(len(vertices_list)):
        for j in range(i,len(vertices_list)):
            if vertices_list[i] == vertices_list[j] and vertices_list[j] not in traingles:
                traingles.append(vertices_list[i])
    #print(traingles)
    count = len(traingles)
    # Return the number of triangles found
    return count
```
